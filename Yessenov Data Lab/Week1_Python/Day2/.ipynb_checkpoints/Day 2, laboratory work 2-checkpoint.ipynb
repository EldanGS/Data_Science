```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import math
%matplotlib inline
```

```python
x = np.linspace(0, math.pi * 2, 10)
y = x
plt.plot(x, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.7, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.7, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
axes1.plot(x, y, color = 'red')
```

```python
x = np.array([1, 5, 10, 15, 20, 25, 30])
y = x ** 2

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.7, 0.8])
axes1.plot(x, y, color = 'red')
plt.xlabel('X')
plt.ylabel('Y')

axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
x = np.array([10, 10.1, 10.2, 10.3])
y = x + 0.1
axes2.plot(x, y + 5, color = 'blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('zoom')
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'b--')
```

```python
x = np.linspace(0, math.pi * 2, 10)
y = x
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')
y = x ** 4
plt.subplot(2, 2, 2)
plt.plot(x, y, 'b')
y = x
plt.subplot(2, 2, 3)
plt.plot(x, y, 'y')
y = x ** 2
plt.subplot(2, 2, 4)
plt.plot(x, y, 'g--')
```

```python
x = np.linspace(0, math.pi * 2, 50)
y = np.sin(x)

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 1, 1])
axes2 = fig.add_axes([0.2, 0.3, 0.2, 0.2])
axes3 = fig.add_axes([0.6, 0.6, 0.3, 0.3])

axes1.plot(x, y, color = 'red')
axes3.plot(x, -y, color = 'blue')
```
