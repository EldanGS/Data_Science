```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fbprophet import Prophet

%matplotlib inline
```

## Load data

```python
df = pd.read_csv('ts.csv')
df.head()
```

## Data review

```python
df.info()
```

```python
df.describe()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df.info()
```

```python
df['ds'] = df['DATE']
df.drop('DATE', axis = 1, inplace = True)
```

## Model splits

```python
df_in_for = df[['ds', 'IN_FOR']]
df_in_for['y'] = df['IN_FOR']
df_in_for.drop('IN_FOR', axis = 1, inplace = True)

df_out_for = df[['ds', 'OUT_FOR']]
df_out_for['y'] = df['OUT_FOR']
df_out_for.drop('OUT_FOR', axis = 1, inplace = True)

df_in_kzt = df[['ds', 'IN_KZT']]
df_in_kzt['y'] = df['IN_KZT']
df_in_kzt.drop('IN_KZT', axis = 1, inplace = True)

df_out_kzt = df[['ds', 'OUT_KZT']]
df_out_kzt['y'] = df['OUT_KZT']
df_out_kzt.drop('OUT_KZT', axis = 1, inplace = True)
```

## Build model

### Model IN_FOR

```python
m = Prophet()
m.fit(df_in_for)
```

```python
# Future
future = m.make_future_dataframe(periods=180)
future.tail()
```

```python
# The forecast object here is a new dataframe that includes a column yhat with the forecast, 
# as well as columns for components and uncertainty intervals.
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```
