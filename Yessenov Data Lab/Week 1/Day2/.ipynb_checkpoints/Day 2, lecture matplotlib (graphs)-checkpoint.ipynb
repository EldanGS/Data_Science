```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import math
%matplotlib inline
```

```python
x = np.linspace(0, math.pi * 2, 10)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r--')
```

```python
plt.plot(x, y, 'g*-')
```

```python
plt.plot(x, y, 'r')
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('Graph title')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'g*-')
plt.subplot(2, 2, 3)
plt.plot(x, y, 'b--')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'y*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.7, 0.5, 0.4, 0.3])
axes1.plot(x, y, 'r')
axes2.plot(x, -y, 'b')
axes2.set_ylabel('-sin')
fig.savefig('twosines.png', dpi = 600)
```

```python
plt.plot(x, y)
plt.plot(x + 1, -y, color = "g")
```

```python
plt.plot(x, y, color=matplotlib.colors.rgb2hex((0, 0, 0)))
plt.plot(x + 1, -y, color = 'r')
```

```python
plt.plot(x, y, color = 'purple', lw = 2, ls = '-', marker = 'o', markersize = 3)
plt.plot(x, y + 0.2, color = 'green', lw = 2, ls = '-', marker = '*', markersize = 4)
plt.plot(x, y + 0.4, color = 'yellow', lw = 2, ls = '-', marker = 'v', markersize = 5)
plt.plot(x, y + 0.6, color = 'red', lw = 2, ls = '-', marker = '^', markersize = 6)
```

```python
plt.scatter(x, y, c = [matplotlib.colors.rgb2hex((1 - 1 / (i + 1), 0, 1 / (i + 1))) for i in range(10)])
```
