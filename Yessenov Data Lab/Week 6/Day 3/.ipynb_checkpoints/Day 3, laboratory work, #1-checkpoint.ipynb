```python
import pandas as pd
import numpy as np
import seaborn as sns
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt

%matplotlib inline
```

### Load data

```python
data = pd.read_excel('data.xlsx')
```

```python
data.head()
```

```python
data = data[data.columns[0:3]]
data.info()
```

### Clean data

```python
data['CARS'] = data['HEAD_ANNOUNCEMENT'].apply(lambda x: x.split(' ')[0])
```

```python
data['CARS']
```

```python
data['CARS_YEAR'] = data['HEAD_ANNOUNCEMENT'].apply(lambda x: x.split(' ')[-2])
data['CARS_YEAR']
```

```python
data.info()
```

```python
data.head().T
```

```python
group_cars = data[data['CARS_YEAR'] >= '2005'].pivot_table(values = 'ID', index = 'CARS', aggfunc = 'count')
group_cars
```

```python
group_cars.head()
```

```python
plt.pie(group_cars, shadow = False)
```
