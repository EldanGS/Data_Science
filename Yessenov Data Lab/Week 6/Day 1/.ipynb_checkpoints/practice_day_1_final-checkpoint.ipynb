```python
import numpy as np
```

```python
data['TARGET']=np.where(data['IS_WALLET_USER'].notnull(),1,0)
```

```python
data.info()
```

```python
data=data.drop(['IS_WALLET_USER','ISWALLETUSER', 'IDATE', 'USER_ID'],1)
```

```python
data.info()
```

```python
print (data['TARGET'].value_counts())
```

```python
# Split to train and test
X_dataset = data.ix[:, data.columns != 'TARGET']
y_dataset = data.ix[:, data.columns == 'TARGET']
X_train, X_test, y_train, y_test = train_test_split(X_dataset, y_dataset,
                                                        test_size=0.3,
                                                        random_state=12)
```

```python
print ('Number of TRAIN set is', len(X_train))
print ('Number of TEST set is', len(X_test))
```

```python
data=data.drop_duplicates()
ProfileReport(data)
```

```python
'''
    RandomForest
'''

rf = RandomForestClassifier(n_estimators=500)
rf.fit(X_train, y_train['TARGET'].values)
predicted = rf.predict(X_test)
```

```python
print ('Random Forest F1 score is', f1_score(y_test['TARGET'].values, rf.predict(X_test),average='binary'))
print ('Random Forest precision is', precision_score(y_test['TARGET'].values, predicted, average='binary'))
print ('Random Forest recall is', recall_score(y_test['TARGET'].values, predicted, average='binary'))
```

```python
predicted_probability_test = [x[1] for x in rf.predict_proba(X_test)]
df_test_result = pd.DataFrame({'TARGET': y_test['TARGET'].values, 'PREDICTED': predicted_probability_test})
```

```python
# Test
df_green = df_test_result.loc[df_test_result['PREDICTED'] < 0.2, :]
print ('Green zone')
print (df_green['TARGET'].value_counts())
df_yellow = df_test_result.loc[(df_test_result['PREDICTED'] >=0.2) & (df_test_result['PREDICTED'] <= 0.7), :]
print ('Yellow zone')
print (df_yellow['TARGET'].value_counts())
df_red = df_test_result.loc[df_test_result['PREDICTED'] > 0.7, :]
print ('Red zone')
print (df_red['TARGET'].value_counts())
```
